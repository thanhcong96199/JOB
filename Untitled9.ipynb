```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import re

def getInfomation():#hàm cạo dữ liệu

    job_title, job_company, job_links,  job_locations,  job_date  = [],[], [], [], [] # khai báo các biến để lưu hết dl của 14 trang
    for i in range(1,15):
        get_page = 'trang' +'-'+ str(i) # Tạo ra từng trang 1 đến trang 14
        urls = 'https://careerbuilder.vn/viec-lam/cntt-phan-mem-c1l4-' + get_page + '-vi.html' # truyền get_page vào url để lấy requests
        req = requests.get(urls)
        soup = BeautifulSoup(req.text, 'html.parser')# Lấy ra file text của các các urls

        tag_h3 =  soup.find_all('h3',{'class': 'job'}) # Tìm thẻ h3 có class là job
        job_title.append([tag_h3[i].text for i in range(0,len(tag_h3))]) # Lấy text chứa trong h3 xong rồi đẩy vào job_titile, job_title sẽ trả 1 list

    
        tag_p_namecom = soup.find_all('p', {'class': 'namecom'}) # Tìm thẻ p có class là namecom
        job_company.append([tag_p_namecom[i].text for i in range(0, len(tag_p_namecom))]) # Lấy text chứa trong thẻ p rồi đẩy vào job_company, trả về list

        job_links.append([tag_h3[i].a.get('href') for i in range(0,len(tag_h3))]) # Lấy links thẻ a nằm trong thẻ h3 xong lưu vào job_links

        tag_p_location = soup.find_all('p',{'class': 'location'}) # Tìm tất cả thẻ p có class: location
        job_locations.append([tag_p_location[i].text for i in range(0, len(tag_p_location))]) # Lấy text của thẻ p rồi lưu vào trong job_locations
        tag_div_dateposted = soup.find_all('div', {'class': 'dateposted'}) # Tìm thẻ div có class là dateposted
        job_date.append([tag_div_dateposted[i].text for i in range(0, len(tag_div_dateposted))]) # Lấy text của thẻ div lưu trong job_date
        
    return job_title, job_company,job_links,  job_locations, job_date




get_job_title, get_job_company, get_job_links, get_job_locations, get_job_date = getInfomation() # Tạo ra 5 biến để lưu giá trị trả về của hàm getInfomation()
# Vì get_job_title sẽ trả về kiểu  list dạng như sau: [ [], [], [], ....,[]] nên ta phải lấy từng list con trong get_job_title thì sẽ thuận tiện cho việc sẽ bảng
# Tương tự đối với get_job_company, get_job_links, get_job_locations, get_job_date
def getJobTitle():
    job_title = []
    for i in range(0, len(get_job_title)):
        job_title += get_job_title[i] # giống như job_title = get_job_title[0] + get_job_title[1] + ...+get_job_title[13] vì có tất cả 14 trang
    return job_title

job_title = getJobTitle()

def getJobCompany():
    job_company = []
    for i in range(0, len(get_job_company)):
        job_company += get_job_company[i]
    return job_company
job_company = getJobCompany()

def getJobLink():
    job_links = []
    for i in range(0, len(get_job_links)):
        job_links += get_job_links[i]
    return job_links
job_links = getJobLink()

def getJobLocation():
    job_locations = []
    for i in range(0, len(get_job_locations)):
        job_locations += get_job_locations[i]
    return job_locations
job_locations = getJobLocation()


def getJobDate():
    job_date = []
    for i in range(0, len(get_job_date)):
        job_date += get_job_date[i]
    return job_date
job_date = getJobDate()

# Tạo pd.DataFrame để vẽ bảng
data = pd.DataFrame({'Title': job_title, 'Company':job_company, 'Link': job_links, 'Location': job_locations, 'Date': job_date})
# Vì mặc định số dòng(rows) trong DataFrame là 60 và số cột(columns) là 20. Mà tổng dữ liệu của chúng ta là 700 dòng nên ta phải lấy số dòng >= 700
# Câu lệnh lấy số dòng là:
pd.options.display.max_rows = 701


# Giờ gọi data lên để nhìn thấy kết quả
#print(data)

# Tạo ra hàm lấy ngày tháng năm, với dữ liệu truyền vào là ngày dưới định dạng day/month/year

def getMonthDay(agrv):
    test_str = agrv 
    regex = r"/(.*?)/"  # Định dạng chuỗi so khớp để lấy tháng
    regexOne = r"^\d+" # Định dạng chuỗi so khớp để lấy ngày
    
    matches = re.finditer(regex, test_str) # lấy địa chỉ
    matchesOne = re.finditer(regexOne, test_str) #lấy địa chỉ
    # Lấy được tháng 
    for matchNum, match in enumerate(matches):
        matchNum = matchNum + 1
        for groupNum in range(0, len(match.groups())):
            groupNum = groupNum + 1
            group = match.group(groupNum)
            
    # Lấy được ngày
    for matchNumOne, matchOne in enumerate(matchesOne):
        matchOne = matchOne.group()
    
    return group, matchOne # Trả về tháng, ngày


# Lấy ngày: vì dữ liệu truyền vào có dạng 'Ngày cập nhập: day/month/year. Để lấy mỗi day/month/year

def getDate(agrv):
    regex = r"\b(0?[1-9]|[12]\d|3[01])[\/\-.](0?[1-9]|[12]\d|3[01])[\/\.-](\d{2}|\d{4})\b"
    test_str = agrv
    matches = re.finditer(regex, test_str)
    for mathNum, match in enumerate(matches):
        match = match.group()
        month, day = getMonthDay(match)
        
        
    return match, month, day


list_dates, list_monthes, list_days = [], [], []
# Hàm này để lấy ngày, và số lượng ngày 
def monthDay():
    listCountDay = []
    listDay = []
    for date in job_date:

        firt, second, thirt = getDate(date)
        list_dates.append(firt)
        list_monthes.append(second)
        list_days.append(thirt)

    # Câu lệnh: Lấy ra các ngày không bị trùng lặp 
    listDayNotDuplicates = list(set([list_day for list_day in list_days if list_days.count(list_day) >= 1]))
    # Câu lệnh: Sắp xếp lại các ngày không  bị trùng lặp theo thứ tự từ nhỏ đến lớn, note: duplicates: không trùng lặp
    listDayNotDuplicates.sort(reverse = False)
    
  
    # Vòng for này để đếm số lượng ngày xuất hiện trong list_days
    # Tạo định dạng ngày như câu lệnh day =...
    for listDayNotDuplicate in listDayNotDuplicates:
        count = list_days.count(listDayNotDuplicate)
        listCountDay.append(count)
        day = 'Ngay {date}'.format(date = listDayNotDuplicate)
        listDay.append(day)
 

    return listDay, listCountDay
# Hàm lấy số lượng từng địa điểm tuyển việc

def getJobLocation():
    listJobLocation, countJobLocations = [], []
    for job_location in job_locations: # Vòng for này để lấy 1 list chứa các địa điểm theo dạng 'hà nội', 'hồ chí minh' không giống như job_locations ban đầu có cả 'Hà nội, hồ chí minh, đà nẵng', 'hà nội'
    
        listJobLocation += job_location.split(',')
    # 2 câu lệnh này có nghĩa chỉ lấy địa điểm không bị trùng vd như 'hà nội', 'hà nội' chỉ tính 'hà nội' trong list lịtobLocation
    getListLocations = list(set([getListLocation for getListLocation in listJobLocation if listJobLocation.count(getListLocation) >=1 ]))
    getListLocations.sort(reverse = False) # sắp xếp theo thứ tự anpha bêta
    
    # Vòng for này để đếm số lượng xuất hiện của địa điểm
    for getListLocation in getListLocations:
        countJobLocation = listJobLocation.count(getListLocation)
        countJobLocations.append(countJobLocation)
        
    return getListLocations, countJobLocations
        

Day, countDay = monthDay()
nameJobLocation, countJobLocation = getJobLocation()
tableDay = pd.DataFrame({'Ngay': Day, 'Count': countDay})
tableLocation = pd.DataFrame({'Name Location': nameJobLocation, 'Count Location': countJobLocation})
# Ve bieu do phan bo so luong cong viec trong thang 5
tableDay.plot.bar(x = 'Ngay', color = 'red')
plt.xlabel('Ngay')
plt.ylabel('Count')
plt.title('Bieu do phan bo viec lam trong thang 5')

# Ve bieu do phan bo dia diem tuyen viec lam
tableLocation.plot.bar(x = 'Name Location', color = 'red')
plt.xlabel('Ten Dia Diem Tuyen Viec')
plt.ylabel('So Luong Dia Diem')
plt.title('Bieu do phan bo viec lam theo khu vuc tuyen dung')

    
```
